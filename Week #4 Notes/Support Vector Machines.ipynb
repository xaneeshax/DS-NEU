```python
from sklearn.svm import LinearSVC
```

```python
import pandas as pd
from sklearn.datasets import load_digits

# load the digits dataset
digits = load_digits()

df = pd.DataFrame(digits.data)
df['target'] = digits.target

df.head()
```

```python
features = df.drop('target', axis = 1)
target = df['target']

features.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state = 3000)

# Select a Classifier
# max_iter : this is related to the linear classifier
# if you don't specify this you get an error message
clf = LinearSVC(random_state = 3000, max_iter = 1000000)

# Create the model by fitting the training data
clf.fit(X = X_train, y = y_train)

# Prediction Accuracy
accuracy = clf.score(X_test, y_test)

print('Prediction accuracy on the test data', f'{accuracy: .2%}')
```
